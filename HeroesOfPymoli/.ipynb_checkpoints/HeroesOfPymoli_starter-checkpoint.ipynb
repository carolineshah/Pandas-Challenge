```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
# retrieve number of unique players
total_players = purchase_df["SN"].nunique()
print("Total number of players: " + str(total_players))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
summary_df = pd.DataFrame()
# find all unique items then count how many
summary_df["Unique items"] = purchase_df["Item Name"].value_counts()
summary_df["Number of unique items"] = summary_df["Unique items"].count()

summary_df["Average purchase price"] = purchase_df["Price"].mean()
summary_df["Total number of purchases"] = purchase_df["SN"].count()
summary_df["Total revenue"] = purchase_df["Price"].sum()
# Formatting
summary_df["Average purchase price"] = summary_df["Average purchase price"].map("${:.2f}".format)
summary_df["Total revenue"] = summary_df["Total revenue"].map("${:.2f}".format)
summary_a_df = summary_df[["Number of unique items", "Average purchase price", 
                            "Total number of purchases", "Total revenue"]]
summary_a_df.reset_index(drop=True, inplace=True)
summary_a_df.drop_duplicates(inplace=True)
summary_a_df.head()
```

```python
purchasing_df = pd.DataFrame()
grouped_df = purchase_df.groupby(["Gender"])
purchasing_df["Purchase Count"] = purchase_df["Gender"].value_counts()
purchasing_df["Average Purchase Price"] = grouped_df["Price"].mean()
purchasing_df["Total Purchase Value"] = grouped_df["Price"].sum()
purchasing_df["Average Purchase Total per Person"] = purchasing_df["Total Purchase Value"] / grouped_df["SN"].nunique()

purchasing_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


Fix formatting

```python
age_df = pd.DataFrame()
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
bin_names = ["<10", "10-14", "15-19", "20-24",
          "25-29", "30-34", "35-39", "40+"]
purchase_df["Age Labels"] = pd.cut(purchase_df["Age"], bins, labels=bin_names)
grouped_age = purchase_df.groupby(["Age Labels"])
age_df["Total Count"] = grouped_age["SN"].nunique()
age_df["Percentage of Players"] = age_df["Total Count"] / age_df["Total Count"].sum() * 100
age_df

```

```python
top_spenders_df = pd.DataFrame()
grouped_spend = purchase_df.groupby(["SN"])
top_spenders_df["Purchase Count"] = grouped_spend["SN"].count()
top_spenders_df["Total Purchase Value"] = grouped_spend["Price"].sum()
top_spenders_df.sort_values(by=["Total Purchase Value"],ascending=False, inplace=True)
top_spenders_df["Average Purchase Price"] = top_spenders_df["Total Purchase Value"] / top_spenders_df["Purchase Count"] 
top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = pd.DataFrame()
# popular_df["Item ID"] = purchase_df["Item ID"]
# popular_df["Item Name"] = purchase_df["Item Name"]
# popular_df["Price"] = purchase_df["Price"]

grouped_items = purchase_df.groupby(["Item ID", "Item Name", "Price"])
popular_df["Purchase Count"] = grouped_items["Item Name"].count()
popular_df.reset_index(level=["Price"], inplace=True)
popular_df["Total Purchase Value"] = popular_df["Purchase Count"] * popular_df["Price"]
popular_df.sort_values(by=["Purchase Count"],ascending=False, inplace=True)
popular_df
```
